```python
import matplotlib.pyplot as plt
x_values=list(range(-4,5))
y_values=[x**2 for x in x_values]
plt.plot(x_values,y_values)
plt.title("f(x)=x^2",fontsize=20)
plt.tick_params(axis="both",which="both",labelsize=12)
plt.axis(xmin=-7, xmax=7 )
plt.xlabel('X',fontsize=12)
plt.ylabel('X^2',fontsize=12)
plt.show()
```

```python
from pandas import read_csv
from matplotlib import pyplot
from matplotlib.pyplot import MultipleLocator


series = read_csv('data.csv', header=0, index_col=0)
pyplot.plot(series)
#pyplot.tick_params(axis="both",which="both",labelsize=12)
#pyplot.axis(xmin=1542003540, xmax=1542608280)
#pyplot.axis(xmin=-1, xmax=1542608280)
# Locator attempting to generate 6049 ticks ([1542003500.0, ..., 1542608300.0]), which exceeds Locator.MAXTICKS (1000).
#x_major_locator=MultipleLocator(100) 
#y_major_locator=MultipleLocator(100)
x_major_locator=MultipleLocator(100000) 
y_major_locator=MultipleLocator(100000)
ax=plt.gca()
ax.xaxis.set_major_locator(x_major_locator) 
ax.yaxis.set_major_locator(y_major_locator) 
pyplot.xlim(1542003540,1542608280)
plt.ylim(-1000000,1000000)
pyplot.show()
    
```
